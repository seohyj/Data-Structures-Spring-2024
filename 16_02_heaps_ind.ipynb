### Simple Python List (Slow)

```python
class PriorityQueue():
  def __init__(self):
    self._data = [] # 리스트 이용할꺼니까 리스트로

  def insert(self, x):
    self._data.append(x)

  def findmin(self):
    if len(self._data) == 0:
      raise ValueError("empty priority Queue")
    # else
    return min(self._data) # O(N); not sorted items / but we want O(log N)

  def deletemin(self):
    # 1. find the min (value and its position)
    # 2. delete the min value from the list
    # 1.
    min_val = self.findmin()
    min_idx = self._data.index(min_val) # .index는 자리를 반환하는 기본 내장함수
    # 2.
    self._data.pop(min_idx)
    return min_val
```

### Sorted Python List

위의 방법은 비효율적임으로 sorted python list를 이용해서 구현

--> 파이썬의 리스트에 아이템을 저장할때, 아이템을 sorting 하여 python list에서 priority 구현하는 것

```python
class PriorityQueue():
  def __init__(self):
    self._data = []

  def insert(self, x): # 앞선 구현에서는 insert가 그냥 맨 뒤에 붙이면 되서 간단한 대신 다른 impl이 복잡.
    # 이게 조금 복잡해지지만 전체적인 복잡도를 분산, 낮춰줌
    # put smallest item at the end of the list
    i = 0
    # i가 리스트 끝에 도달하기 전, 그리고 x가 현재 위치의 값보다 작다면
    while i < len(self._data) and x < self._data[i]:
      #x를 1 증가
      i += 1
    self._data.insert(i, x) # i에 x를 삽입

  def findmin(self):
    return self._data[-1] # O(1) complexity

  def deletemin(self):
    return self._data.pop() # O(1) complexity
```

findmin, deletemin이 더 간단함

상황에 따라 findmin, deletemin에 많이 쓰이게 될 경우 sorted python list로 구현 / insert가 많이 구현되는 상황으로 쓰일 경우 위의 sol1 / sol 2로 구현

### Binary Heap

두가지 방법이 있음
- Linked List
- Arrays

수업에서는 Array 이용해서 구현; 왜냐하면 Binary Tree 가  complete한 상황을 고려하기 때문

Of course a complete binary tree can be implemented as a linked data structure, with Node objects that store references to their parent and children.

However, because the tree is complete, we can actually store it very compactly in an array or Python list. To do so, simply number the nodes of the tree from top to bottom, left to right, starting with one for the root, so that the nodes are numbered from 1 to n.

We observe that the left child of node i has index 2i and the right child has index 2i + 1. The parent of node i has index ≈(i/2). We will simply store node i in slot i of an array, and we can move from a node to its parent or its children by doing simple arithmetic on the index. (We started numbering from 1 for convenience—the numbers are nicer. Of course this means that we waste slot 0 of the array—not a big problem, but it could be fixed by changing the numbering scheme slightly.)

Complete Binary Queue 에서의 관계식:

- 부모 노드를 p 라고 할 때, (parent = p)
 - left_child = 2 * p
 - right_child - 2 * p + 1

- 자식 노드를 g 라고 할 때, (child = g)
 - parent = g // 2

또한 일반적으로 노드의 value로는 0을 쓰지 않음

```python
class PriorityQueue():
  DEFAULT_CAPACITY = 100

  def __init__(self):
    self._data = [ None ] * PriorityQueue.DEFAULT_CAPACITY
    self._size = 0

  def __len__(self):
    return self._size

  def findmin(self):
    return self._data[1] # return root node; O(1). very efficient!

  def insert(self, x):   # x를 넣는 것.
  # heap order 충족을 위해 bubble up,
  # complete 상태를 유지하기 위해 항상 leftmost 자리에 채우고 꽉 차면 다음 Level로 넘어감
    if self._size + 1 == len(self._data):         # data가 꽉 찼다면,
      self._data.extend([None] * len(self._data)) # 길이를 2배로 늘려줌
    self._size += 1           # size를 1 늘려줌,
    hole = self._size         # size의 위치는 hole이 됨,
    # 이때 item을 바로 넣는게 아니라 적절한 위치인지 찾는 bubble up 구현

    # bubble up: 나누기 2 operation으로 array에서 왔다갔다 구현
    while x < self._data[hole // 2]: # x가 parent 값보다 작으면,
      self._data[hole] = self._data[hole//2] # swap
      hole //=2                              # move-up
    self._data[hole] = x
    # x를 넣어서 미리 바꾸는게 아니라, hole의 위치에 위에꺼를 넣어놓고 나머지 자리에 x를 넣는 것

  def deletemin(self):
    min_item = self.findmin()
    self.data[1] = self._data[self._size] # 끝의 item을 넣어줌
    self._size -= 1
    self._bubble_down(1)
    return min_item

  def _bubble_down(self, i):
    value = self._data[i]
    hole = i
    child = self._smaller_child(hole, value)
    while child != 0:
      self._data[hole] = self._data[child]
      hole = child
      child = self._smaller_child(hole, value)
    self._data[hole] = value

  # smaller child를 찾는 과정이 필요함
  def _smaller_child(self, index, value):
    child = 2 * index # left child
    if child <= self._size: # True인 경우 child가 있다는 뜻
      if child != self._size and self._data[child+1] < self._data[child]
      # 끝에 위치한게 아니거나 / 오른쪽 child와 비교해 오른쪽 child가 더 작으면
        child += 1
      if self._data[child] < value:
        return child
    # else
    return 0
```

**Approach 1**
1. Construct a Max Heap.
2. Add all elements into the Max Heap.
3. Traversing and deleting the top element (using pop() or poll() for instance).
4. Repeat Step 3 K times until we find the K-th largest element.

**Approach 2**
1. Construct a Min Heap with size K.
2. Add elements to the Min Heap one by one.
3. When there are K elements in the “Min Heap”, compare the current element with the top element of the Heap:
  - If the current element is not larger than the top element of the Heap, drop it and proceed to the next element.
  - If the current element is larger than the Heap’s top element, delete the Heap’s top element, and add the current element to the “Min Heap”.
4. Repeat Steps 2 and 3 until all elements have been iterated.

```python
import heapq

def kth_smallest(nums, k):
  heap = []
  for num in nums:
    heapq.heappush(heap, num)  # insert
  # heapq.heapify(nums)

  kth_min = None
  for _ in range(k):
    kth_min = heapq.heappop(heap)  # deletemin
  return kth_min

print(kth_smallest([4, 1, 7, 3, 8, 5], 3))
print(heapq.nlargest(4, [4, 1, 7, 3, 8, 5]))
print(heapq.nsmallest(3, [4, 1, 7, 3, 8, 5]))
```
