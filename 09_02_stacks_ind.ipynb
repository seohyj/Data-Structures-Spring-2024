## Implementing a Stack

```python
import doctest

class EmptyStackError(Exception):  # when stack is empty <= pop()
  pass

class StackOverflow(Exception):  # when stack is full <= push(item)
  pass
```

### Using the Python List

The Python list makes our life simple.

```python
class Stack():
  def __init__(self):
    self._data = []
    # [1, 2, 3, 4, 5]
    #  b           t

  def is_empty(self):
    return len(self._data) == 0

  def push(self, item):
    self._data.append(item)

  def top(self):
    if self.is_empty():
      raise EmptyStackError
    return self._data[-1]

  def pop(self):
    if self.is_empty():
      raise EmptyStackError
    return self._data.pop()

  def size(self):
    return len(self._data)
```

```python
class Stack():
  def __init__(self):
    self._data = []

  def is_empty(self):
    return len(self._data) == 0

  def push(self, item):
    self._data.append(item)

  def top(self):
    if self.is_empty(): # check if the stack is empty or not
      raise EmptyStackError
    return self._data[-1]

  def pop(self):
    if self.is_empty(): # check if the stack is empty or not
      raise EmptyStackError
    return self._data.pop() # return last item (top item)

  def size(self):
    return len(self._data)
```

```python
stack = Stack()
print('after we initialized a stack:', stack.is_empty())
stack.push(5)
print('after pushing 5 into the stack:', stack.is_empty())
stack.push(8)
stack.push(9)
print('the top of the stack:', stack.top())
print('popping from the stack:', stack.pop())
print('popping from the stack:', stack.pop())
print('popping from the stack:', stack.pop())
print('after popping three times from the stack:', stack.is_empty())
```

### Using Arrays

However, you might not be able to use Python all the times. You sometimes need to write code in other languages such as C, C++ or Java. In such cases, we implement a Stack with arrays.

```python
class Stack:
    def __init__(self, capacity):
        self._data = [None] * capacity  # Initialize a list with 'capacity' number of None elements
        self._tos = -1  # Initialize the top of stack pointer to -1 (empty stack)

    def is_empty(self):
        return self._tos == -1  # Check if the stack is empty by comparing top of stack pointer to -1

    def push(self, item):
        if self._tos + 1 == len(self._data):
            raise StackOverflow  # Raise an exception if the stack is full (top of stack pointer at end of list)
        self._tos += 1  # Increment the top of stack pointer
        self._data[self._tos] = item  # Add the item to the top of the stack

    def top(self):
        if self.is_empty():
            raise EmptyStackError  # Raise an exception if the stack is empty
        return self._data[self._tos]  # Return the top element of the stack

    def pop(self):
        if self.is_empty():
            raise EmptyStackError  # Raise an exception if the stack is empty
        item = self._data[self._tos]  # Get the top element of the stack
        self._tos -= 1  # Decrement the top of stack pointer
        return item  # Return the top element of the stack
```

```python
stack = Stack()
print('after we initialized a stack:', stack.is_empty())
stack.push(5)
print('after pushing 5 into the stack:', stack.is_empty())
stack.push(8)
stack.push(9)
print('the top of the stack:', stack.top())
print('popping from the stack:', stack.pop())
print('popping from the stack:', stack.pop())
print('popping from the stack:', stack.pop())
print('after popping three times from the stack:', stack.is_empty())
```

Implementing min_stack in Arrays NOT LISTS

```python
class MinStack(object):
    def __init__(self):
        self.stack = []          # Initialize the stack as an empty array
        self.min_stack = []      # Initialize the min_stack as an empty array

    def push(self, item):
        self.stack.append(item)                # Add the item to the top of the stack
        if not self.min_stack or self.min_stack[len(self.min_stack) - 1] >= item:
            self.min_stack.append(item)
        # If the min_stack is empty or the new item is smaller than or equal to the top of min_stack,
        # add it to min_stack

    def top(self):
        return self.stack[len(self.stack) - 1]
        # Return the top element of the stack

    def pop(self):
        tmp = self.stack.pop(len(self.stack) - 1)     # Remove and return the top element of the stack
        if tmp == self.min_stack[len(self.min_stack) - 1]:
            self.min_stack.pop(len(self.min_stack) - 1)
            # If the popped element is the same as the top element of min_stack, remove it from min_stack
        return tmp

    def get_min(self):
        return self.min_stack[len(self.min_stack) - 1]
        # Return the top element of min_stack, which represents the current minimum element in the stack
```

Let's break down the condition not self.min_stack or self.min_stack[len(self.min_stack) - 1] >= item:

not self.min_stack: This part of the condition checks if self.min_stack is empty. If self.min_stack is empty, the expression not self.min_stack evaluates to True.
self.min_stack[len(self.min_stack) - 1] >= item: This part of the condition checks if the top element of self.min_stack (i.e., the last element of the array) is greater than or equal to the new item being pushed onto the stack.
Now, let's put it all together:

If self.min_stack is empty (not self.min_stack evaluates to True), it means there are no minimum elements stored in min_stack yet. In this case, the condition evaluates to True.
If self.min_stack is not empty, the condition checks if the top element of self.min_stack is greater than or equal to the new item. If it is, it means the new item is smaller than or equal to the current minimum element in the stack. In this case, the condition evaluates to True.
Overall, this condition is used to determine whether the new item should be added to self.min_stack based on whether it's smaller than or equal to the current minimum element in the stack. If either self.min_stack is empty or the new item is smaller than or equal to the top element of self.min_stack, the condition evaluates to True, and the new item is added to self.min_stack.

Implement Stack using the `_Node` class

```python
class _Node():
  def __init__(self, item, next=None):
    self.item = item
    self.next = next

class Stack():
  def __init__(self):
    self._tos = None

  def is_empty(self):
    return self._tos is None

  def push(self, item):
    self._tos = _Node(item, self._tos)

  def top(self):
    if self.is_empty():
      raise EmptyStackError
    return self._tos.item

  def pop(self):
    if self._is_empty():
      raise EmptyStackError
    item = self._tos.item
    self._tos = self._tos.next
    return item
```

```python
stack = Stack()
print('after we initialized a stack:', stack.is_empty())
stack.push(5)
print('after pushing 5 into the stack:', stack.is_empty())
stack.push(8)
stack.push(9)
print('the top of the stack:', stack.top())
print('popping from the stack:', stack.pop())
print('popping from the stack:', stack.pop())
print('popping from the stack:', stack.pop())
print('after popping three times from the stack:', stack.is_empty())
```

### Checking Balanced Parenthesis

Our job is to check whether an arithmetic expression is correctly balanced.

For instance,
* (){[(){}]([])} is correctly balanced.
* (){[({)}]([])} is not because a round ) closes a curly { bracket.

We use the following strategy

1. Make an empty stack,
2. For each symbol in the string:

  (a) If the symbol is an opening symbol, push it on the stack.

  (b) If it is a closing symbol, then

        i. If the stack is empty, return false.
        ii. If the top of the stack does not match the closing symbol, return false.
        iii. Pop the stack.
3. Return true if the stack is empty, otherwise false.

Below is the implementation

```python
def matching(open, close):
  m = { '(' : ')', '{': '}', '[': ']' }
  return open in m and m[open] == close

def balanced_symbols(string):
  '''
  >>> balanced_symbols('(){(){}}')
  True
  >>> balanced_symbols('(){({)}}')
  False
  '''
  stack = Stack()

  for ch in string:
    if ch in "({[":
      stack.push(ch)
    elif ch in ")}]":
      if stack.is_empty() or not matching(stack.top(), ch):
        return False
      stack.pop()
    # ignore all other characters
  return stack.is_empty()
```

```python
print(balanced_symbols('(){(){}}'))
print(balanced_symbols('(){({)}}'))
```
